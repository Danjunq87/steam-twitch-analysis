```python
import pandas as pd

# Carrega os datasets
steam = pd.read_csv("../data/games_march2025_cleaned.csv")
twitch = pd.read_csv("../data/sullygnome_top2000.csv")

print(f"Steam: {steam.shape[0]} jogos, {steam.shape[1]} colunas")
print(f"Twitch: {twitch.shape[0]} jogos, {twitch.shape[1]} colunas")
```

```python
# Ver exemplos de categories e genres
print("Exemplo de categories:")
print(steam['categories'].dropna().head(3).values)

print("\nExemplo de genres:")
print(steam['genres'].dropna().head(3).values)

print("\nExemplo de tags:")
print(steam['tags'].dropna().head(3).values)
```

```python
import ast

# Função para converter string para lista/dict
def parse_field(value):
    try:
        return ast.literal_eval(value)
    except:
        return []

# Testa a conversão
exemplo = steam['categories'].dropna().iloc[0]
print("Antes:", type(exemplo), exemplo[:50])
print("Depois:", parse_field(exemplo))
```

```python
# Classifica o tipo de jogo
def classificar_tipo(categories_str, tags_str):
    cats = parse_field(categories_str) if pd.notna(categories_str) else []
    tags = parse_field(tags_str) if pd.notna(tags_str) else {}
    
    cats_str = ' '.join(cats).lower()
    tags_str_lower = ' '.join(tags.keys()).lower() if isinstance(tags, dict) else ''
    combinado = cats_str + ' ' + tags_str_lower
    
    if 'massively multiplayer' in combinado or 'mmo' in combinado:
        return 'MMO'
    elif 'multi-player' in combinado or 'multiplayer' in combinado or 'co-op' in combinado:
        return 'Multiplayer'
    elif 'single-player' in combinado or 'singleplayer' in combinado:
        return 'Single Player'
    else:
        return 'Outro'

steam['tipo'] = steam.apply(
    lambda row: classificar_tipo(row['categories'], row['tags']), axis=1
)

print(steam['tipo'].value_counts())
```

```python
# Verifica se o jogo é Indie
def is_indie(genres_str):
    genres = parse_field(genres_str) if pd.notna(genres_str) else []
    return 'Indie' in genres

steam['indie'] = steam['genres'].apply(is_indie)

print(f"Total de jogos Indie: {steam['indie'].sum()}")
print(f"Total de jogos não-Indie: {(~steam['indie']).sum()}")

print("\nDistribuição por tipo nos jogos Indie:")
print(steam[steam['indie']]['tipo'].value_counts())
```

```python
# Cruza Steam com Twitch pelo nome do jogo
steam_indie = steam[steam['indie']].copy()

# Normaliza os nomes para melhorar o match
steam_indie['name_lower'] = steam_indie['name'].str.lower().str.strip()
twitch_fix['game_lower'] = twitch_fix['game'].fillna('').str.lower().str.strip()

# Faz o cruzamento
merged = steam_indie.merge(twitch_fix, left_on='name_lower', right_on='game_lower', how='inner')

print(f"Jogos Indie encontrados na Twitch: {len(merged)}")
print("\nJogos encontrados:")
print(merged[['name', 'tipo', 'watch_time_hours', 'streamers']].head(20))
```

```python
# Cria faixas de preço
def faixa_preco(price):
    if price == 0:
        return 'Gratuito'
    elif price <= 10:
        return 'Até $10'
    elif price <= 20:
        return 'Até $20'
    elif price <= 30:
        return 'Até $30'
    else:
        return 'Acima de $30'

df_jogos['faixa_preco'] = df_jogos['price'].apply(faixa_preco)

analise_preco = df_jogos.groupby('faixa_preco').agg(
    qtd_jogos=('name', 'count'),
    media_watch_hours=('watch_time_hours', 'mean'),
    media_peak_viewers=('peak_viewers', 'mean')
).round(0).reset_index()

# Ordena pela faixa
ordem = ['Gratuito', 'Até $10', 'Até $20', 'Até $30', 'Acima de $30']
analise_preco['faixa_preco'] = pd.Categorical(analise_preco['faixa_preco'], categories=ordem, ordered=True)
analise_preco = analise_preco.sort_values('faixa_preco')

fig4 = px.bar(
    analise_preco,
    x='faixa_preco',
    y='media_watch_hours',
    color='faixa_preco',
    text='qtd_jogos',
    title='Faixa de Preço vs Média de Horas Assistidas na Twitch — Jogos Indie (2025)',
    labels={
        'faixa_preco': 'Faixa de Preço',
        'media_watch_hours': 'Média de Horas Assistidas',
    }
)

fig4.update_traces(texttemplate='%{text} jogos', textposition='outside')
fig4.update_yaxes(tickformat='.0s')
fig4.write_html("../assets/grafico_preco_faixas.html")
print("Salvo!")
```
